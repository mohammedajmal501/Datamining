**K-mean Cluster**

```python
import numpy as np
from sklearn.cluster import KMeans

X = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_X.npy')
y = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_Y.npy')

# Reshape X to have 2 dimensions
X = X.reshape(X.shape[0], -1)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Evaluate accuracy of clustering
accuracy = np.mean(kmeans.labels_ == y)
print(f"Accuracy of clustering: {accuracy:.2f}")
```

**visualize**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_X.npy')
y = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_Y.npy')

# Reshape X to have 2 dimensions
X = X.reshape(X.shape[0], -1)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Plot clusters
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))

for i in range(3):
    cluster_center = kmeans.cluster_centers_[i]
    cluster_images = X[kmeans.labels_ == i]
    ax[i].imshow(cluster_center.reshape(9, 9), cmap='gray')
    ax[i].set_title(f"Cluster {i+1}\n({len(cluster_images)} images)")
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```

**GMM**

```python
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_X.npy')
y = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_Y.npy')

# Reshape X to have 2 dimensions
X = X.reshape(X.shape[0], -1)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)

# Evaluate accuracy of clustering
accuracy = np.mean(np.argmax(gmm.predict_proba(X), axis=1) == y)
print(f"Accuracy of clustering: {accuracy:.2f}")
```

**Hierarchical clustering**

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_X.npy')
y = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_Y.npy')

# Reshape X to have 2 dimensions
X = X.reshape(X.shape[0], -1)
hclustering = AgglomerativeClustering(n_clusters=3)
hclustering.fit(X)

# Evaluate accuracy of clustering
accuracy = np.mean(hclustering.labels_ == y)
print(f"Accuracy of clustering: {accuracy:.2f}")
```

**Elbow**

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
X = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_X.npy')
y = np.load('/content/drive/MyDrive/Data/smiley_dataset/smiley_Y.npy')

# Reshape X to have 2 dimensions
X = X.reshape(X.shape[0], -1)
# Elbow method
inertias = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)
plt.plot(range(1, 11), inertias)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()

# Silhouette score
scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    scores.append(score)
plt.plot(range(2, 11), scores)
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette score")
plt.show()
```
